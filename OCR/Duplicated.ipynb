```python
import os
import hashlib
from PIL import Image
```

```python
def image_hash(image_path):
    with open(image_path, 'rb') as f:
        image_data = f.read()
        return hashlib.md5(image_data).hexdigest()

def find_duplicates(directory):
    hash_to_path = {}
    duplicates = []

    for root, dirs, files in os.walk(directory):
        for filename in files:
            filepath = os.path.join(root, filename)
            image_hash_value = image_hash(filepath)
            if image_hash_value in hash_to_path:
                duplicates.append((filepath, hash_to_path[image_hash_value]))
            else:
                hash_to_path[image_hash_value] = filepath

    return duplicates

# Replace 'directory' with the path to the directory containing your images
directory = r"C:\Users\Ali\Desktop\DIG\augmnet_v2\0"
duplicates = find_duplicates(directory)

if duplicates:
    print("Duplicates found:")
    for dup in duplicates:
        print(f"Duplicate 1: {dup[0]}, Duplicate 2: {dup[1]}")
else:
    print("No duplicates found.")
```

```python
import os
from PIL import Image
import imagehash

def remove_duplicates(folder_path):
    unique_images = {}  # Dictionary to store unique image hashes
    duplicate_images = []  # List to store duplicate image paths

    # Iterate through the images in the folder
    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        
        # Open the image and calculate its hash
        if image_path != 'C:\\Users\\Ali\\Desktop\\crop\\88\\Thumbs.db':
            with Image.open(image_path) as img:
                img_hash = str(imagehash.average_hash(img))
        
        # Check if the hash is already in the dictionary
        if img_hash not in unique_images:
            unique_images[img_hash] = image_path
        else:
            # If the hash is already in the dictionary, it's a duplicate image
            duplicate_images.append(image_path)

    # Remove duplicate images
    for duplicate_image in duplicate_images:
        #os.remove(duplicate_image)
        print(f'Removed duplicate image: {duplicate_image}')

# Replace 'folder_path' with the path to your dataset folder
folder_path = r'C:\Users\Ali\Desktop\DIG\aug_data\0'
remove_duplicates(folder_path)
```
